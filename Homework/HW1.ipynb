<br/>
It takes an online freely available database with missing entries and calculates how many entries are missing (per column and in the entire dataset), as well as the percentage of values missing per column.</p>

```python
import pandas as pd

# Load the Titanic dataset from an online source
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display the number of missing entries for each column
missing_data = df.isnull().sum()

# Print the total number of missing entries for each column
print("Number of missing entries per column:\n", missing_data)

# Optionally, you can calculate the total missing values in the entire dataset
total_missing = df.isnull().sum().sum()
print(f"\nTotal number of missing entries in the dataset: {total_missing}")

# Display the percentage of missing entries for each column
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nPercentage of missing entries per column:\n", missing_percentage)
```

```python
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Generate summary statistics for each column
summary = df.describe()

# Print the summary statistics
print("Summary Statistics:")
print(summary)

# Generate value counts for each colum
print("\nValue Counts for each column:")
for column in df.columns:
    print(f"\n{column} Value Counts:")
    print(df[column].value_counts())
```

<h1 style="color:red;">Question VIII</h1>

<p>The command df.groupby("col1")['col2'].describe() groups the info of col2 by the unique values of col1. In other words, it takes every unique value from col1 and describes col2 following those values.<br/>For example, in the Titanic database, if we were to use df.groupby('Sex')['Survived'].describe(), we could see survival information for male and female passengers of the Titanic </p>

```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.groupby('Sex')['Survived'].describe()
"""74% of females survived while only 19% of males did"""
```
